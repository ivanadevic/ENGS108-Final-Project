```
#First, we take a large dataset and partition it into geographic subsections.
#For example, if we want data for the roads in Massachusetts, we divide MA into a, say, 1000x2000 grid, defined by (lat,long) coords
#Next, we sort all the features (roads, waters) into their respective sub-tile

#After that, we need to construct an image of each of the (2,000,000 in this case) subtiles.
#We take a list of all the features in that subtile, and try to create some images of them. One image of the roads in that subtile, and one image of the bodies of water

#We can use these two images for each subtiles to machine learn. 
```

```
#First create a sample dataset
#Create some sample features (defined by a list of tuples) (all between (0,0) and (10,10)):
f1 = [(0,1.1), (0.2, 2.0), (0.9, 3), (1.4,6), (2,8)]
f2 = [(0.2,2.0),(1.5,2.1),(3.1,1.8),(6,2),(7,2.1)]
f3 = [(7,2.1),(8.5,3.6),(8.6,4),(9.5,6.7),(10,10)]
f4 = [(7,2.1),(5,1.8),(4.2,1.6),(3.1,1.1),(1.4,0)]
#And bundle them up, into a string that will represent *all of the roads inside of a grid defined by 0<lat<10 and 0<long<10
sample = [f1,f2,f3,f4]
```

```
import matplotlib.pyplot as plt

def prepare_plot(roads_data):
  prepared_dataX = []
  prepared_dataY = []
  for feature in roads_data:
    these_xs = []
    these_ys = []
    for x, y in feature:#apparently this is faster than *zip(): https://stackoverflow.com/questions/8081545/how-to-convert-list-of-tuples-to-multiple-lists
      these_xs.append(x)
      these_ys.append(y)

    #print("prepared_dataX receives", these_xs)
    prepared_dataX.append(these_xs)

    #print("prepared_dataY receives", these_ys)
    prepared_dataY.append(these_ys)
  return prepared_dataX, prepared_dataY

#All done. Now we can graph this stuff

```

```
#Want this function to create a .jpg or .png and save it on my harddrive for each data tuple we pass it...
from matplotlib.pyplot import figure
import numpy as np

def create_image(prepared_dataX, prepared_dataY, i):
  figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k') #Resize to 1000x1000 images
  for i in range(0, len(prepared_dataX)):
    tuple_setX = prepared_dataX[i]
    tuple_setY = prepared_dataY[i]
    for j in range(0, len(tuple_setX), 1):
      plt.plot(tuple_setX[j:j+2], tuple_setY[j:j+2], '-', linewidth = 8, color = 'grey')
      
  plt.axis('off')
  plt.savefig('roads_' +  str(i) + '.jpg', bbox_inches = 'tight')
```

```
#Now we can put the two functions together:

X,Y = prepare_plot(sample)
create_image(X,Y,1)
```
