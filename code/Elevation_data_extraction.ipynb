```
#THIS CODE CHUNK PARSES GEOGRAPHY DATA AND TURNS IT INTO A USEFUL FORMAT
NYC_data_path = "/content/drive/MyDrive/College Year 3/20F/ENGS108/ENGS108_Final_Project/Data_Sets/contours_NYC_test.kml"

from fastkml import  kml

#doc = open(NYC_data_path).read()
#k = kml.KML()
#k.from_string(doc)
#print(doc[0:1000]) #This is just printing the first 1000 characters in the KML document. Not very useful except for organizational purposes


from pykml import parser
import re

root = parser.fromstring(open(NYC_data_path, 'r').read())
sample = []
for i in range(0, len(root.Document.Placemark)):#can do all ~13000 items: len(root.Document.Placemark)
  unprocessed_string = str(root.Document.Placemark[i].LineString.coordinates)
  #print(unprocessed_string)
  items = re.split('[,|\n]', unprocessed_string) #uses regex to split by "," or "\n"
  del items[2::3] #delete zeros
  del items[-1] #delete last item

  processed_string = []
  for i in range(0,len(items),2):
    coord_tup = (float(items[i]), float(items[i+1]))
    processed_string.append(coord_tup)

  sample.append(processed_string)



```

```
import matplotlib.pyplot as plt

def prepare_plot(roads_data):
  prepared_dataX = []
  prepared_dataY = []
  for feature in roads_data:
    these_xs = []
    these_ys = []
    for x, y in feature:#apparently this is faster than *zip(): https://stackoverflow.com/questions/8081545/how-to-convert-list-of-tuples-to-multiple-lists
      these_xs.append(x)
      these_ys.append(y)

    #print("prepared_dataX receives", these_xs)
    prepared_dataX.append(these_xs)

    #print("prepared_dataY receives", these_ys)
    prepared_dataY.append(these_ys)
  return prepared_dataX, prepared_dataY

#All done. Now we can graph this stuff


#Want this function to create a .jpg or .png and save it on my harddrive for each data tuple we pass it...
from matplotlib.pyplot import figure
import numpy as np

def create_image(prepared_dataX, prepared_dataY, i):
  figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k') #Resize to 1000x1000 images
  for i in range(0, len(prepared_dataX)):
    tuple_setX = prepared_dataX[i]
    tuple_setY = prepared_dataY[i]
    for j in range(0, len(tuple_setX), 1):
      plt.plot(tuple_setX[j:j+2], tuple_setY[j:j+2], '-', linewidth = .1, color = 'grey')
      
  plt.axis('off')
  plt.savefig('roads_' +  str(i) + '.jpg', bbox_inches = 'tight')
```

```
X,Y = prepare_plot(sample)
create_image(X,Y,1)
```
